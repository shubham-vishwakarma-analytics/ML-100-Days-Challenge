#  Coding Task

```python
 # Dataset

data = {
    "Hours": [1,2,3,4,5,6,7,8],
    "Practice": [5,10,15,20,25,30,35,40],
    "Result": ['Fail','Fail','Fail','Fail','Pass','Pass','Pass','Pass']
}

import pandas as pd
df = pd.DataFrame(data)

X = df[['Hours', 'Practice']]
y = df['Result']
```

```python
# ✅ Step 1: Import

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
```

```python
# ✅ Step 2: Create Pipeline

pipe = Pipeline([
    ('scaler', StandardScaler()),
    ('model', RandomForestClassifier())
])
```

```python
# ✅ Step 3: Train

pipe.fit(X, y)
```

```python
# ✅ Step 4: Prediction

print(pipe.predict([[6, 30]]))

# 👉 No need to scale manually ❗
# 👉 Pipeline handles everything ✔️
```

## Pipeline + GridSearch (Power Combo)

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'model__n_estimators': [5, 10],
    'model__max_depth': [2, 3]
}

grid = GridSearchCV(pipe, param_grid, cv=3)
grid.fit(X, y)

print("Best Params:", grid.best_params_)
```
