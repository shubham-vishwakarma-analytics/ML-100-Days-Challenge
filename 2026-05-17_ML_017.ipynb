# DAY 17 — Cross Validation (Pro Level)

### Problem with Day 16

#### You did:

* train_test_split → 1 split only

#### 👉 Problem:
* Result depends on how data was split ❌

#### 👉 Maybe:
* One split → lucky → high accuracy
* Another split → low accuracy

**🎯 Solution: Cross Validation**
* 👉 Instead of 1 split:
* We split data multiple times ✔️

### Idea (Very Simple)

**Data → split into parts (folds)**

* Train on some parts  
* Test on remaining part  

* Repeat multiple times  
* → Take average accuracy

#### 🔥 Example (5-Fold CV)
**Split data into 5 parts**

* Round 1 → Test on part 1  
* Round 2 → Test on part 2  
* Round 3 → Test on part 3  
* Round 4 → Test on part 4  
* Round 5 → Test on part 5
* 
**👉 Final:**

#### Average accuracy = final score ✔️

#  Coding Task

```python
# Dataset

data = {
    "Hours": [1,2,3,4,5,6,7,8],
    "Result": [0,0,0,1,1,1,1,1]
}
import pandas as pd
df = pd.DataFrame(data)
df
```

```python
from sklearn.model_selection import train_test_split

X = df[['Hours']]
y = df['Result']
```

```python
# ✅ Step 1: Import

from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsClassifier
```

```python
# ✅ Step 2: Model

model = KNeighborsClassifier(n_neighbors = 3)
```

```python
# ✅ Step 3: Apply Cross Validation

scores = cross_val_score(model, X, y, cv=5)

print(scores)
print("Average Accuracy:", scores.mean())
```
