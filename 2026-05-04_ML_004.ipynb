##  Evaluation Metrics

### Now we go one level deeper:

* 👉 Not just seeing predictions
* 👉 But measuring how good model is



## Concept 

#### We use metrics like:

## 1. MSE (Mean Squared Error)

👉 Measures:

#### How far predictions are from actual values

Formula idea:

(actual - predicted)² → average

**👉 Lower MSE = Better model**

## 2. R² Score (Accuracy for Regression)
👉 Range:

* 0 → bad  
* 1 → perfect

**👉 Closer to 1 = Better model**



## Task 1:

```python
# Answer:
# 1. If prediction error is large → MSE high or low?
# 2. If model is perfect → R² = ?
```

```python
# If error is large → MSE = High ✔️
# 👉 Bigger error → bigger squared value → higher MSE

# If model is perfect → R² = 1 ✔️
# 👉 That means predictions match actual values exact
```

## Coding Time

```python
# Dataset
import pandas as pd
data = pd.DataFrame({'Hours' : [1,2,3,4,5,6,7] , 'Marks' : [12,24,36,48,60,72,84]})
# data
```

```python
# Split the data
from sklearn.model_selection import train_test_split

x = data[["Hours"]]
y = data["Marks"]

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred  = model.predict(X_test)
y_pred
```

##  mean_squared_error, r2_score


```python
# After prediction, add this:
from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(y_test , y_pred)
r2 = r2_score(y_test, y_pred)

print("MSE : ",mse)
print("R2_Score : ",r2)
```

## Task 2:

```python
# If MSE = 0, what does it mean?
```

```python
# It means there is NO error
# → Predicted values = Actual values exactly

# MSE = 0 → Perfect model

# MSE small → Good model

# MSE large → Bad model
```
