#### You give options → Model tests all → Picks best

# Coding Task

```python
 # Dataset

data = {
    "Hours": [1,2,3,4,5,6,7,8],
    "Practice": [5,10,15,20,25,30,35,40],
    "Result": ['Fail','Fail','Fail','Fail','Pass','Pass','Pass','Pass']
}

import pandas as pd
df = pd.DataFrame(data)

X = df[['Hours', 'Practice']]
y = df['Result']
```

```python
# ✅ Step 1: Import

from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
```

```python
# ✅ Step 2: Define Parameters

param_grid = {
    'n_estimators': [5, 10, 20],
    'max_depth': [2, 3, 5]
}
```

```python
# ✅ Step 3: Apply GridSearch

model = RandomForestClassifier()

grid = GridSearchCV(model, param_grid, cv=3)
grid.fit(X, y)
```

```python
# ✅ Step 4: Best Parameters

print("Best Parameters:", grid.best_params_)
print("Best Score:", grid.best_score_)
```

```python

print(f'Best Parameter : {grid.best_params_}' )
print(f'Best Score : {grid.best_score_}' )
```

## What It Does
### 👉 Tries all combinations:

* (5,2), (5,3), (5,5)  
* (10,2), (10,3), (10,5)  
* (20,2), (20,3), (20,5)
  
### 👉 Finds:

* Best combination ✔️

  
## 🧠 Why This is Powerful

* ✔ Saves time  
* ✔ Better performance  
* ✔ No guessing  
* ✔ Used in real-world ML

2. What does cv=3 mean?

```python
# Ans 1 : 
# GridSearchCV is better because it automatically tries multiple parameter combinations,
# uses cross-validation, and finds the best model without manual effort.


# Ans 2: 
# cv = 3 means 3-fold cross validation

# 👉 Meaning:

# * Data is split into 3 parts  
# * Train on 2 parts, test on 1 part  
# * Repeat 3 times  
# * → Average score is calculated ✔️

# 1. GridSearchCV → automatic + uses cross-validation → best model

# 2. cv = 3 → 3-fold cross validation

# 📌 Full Form
# cv = Cross Validation ✔️
```
