## Problem with Single Decision Tree

You learned:

**Decision Tree → can overfit ❌**
* 👉 One tree = risky

**🎯 Solution: Random Forest**
* 👉 Instead of 1 tree:
* Use MANY trees ✔️
👉 Final prediction:

**Majority voting (like KNN idea)**

🌳 Simple Idea

* vTree 1 → Pass  
* vTree 2 → Pass  
* vTree 3 → Fail  
* vTree 4 → Pass
*  
**👉 Final:**

* Pass ✔️ (majority)
* 
**🧠 Why Random Forest is Powerful?**
* ✔ Reduces overfitting  
* ✔ More stable  
* ✔ Better accuracy  
* ✔ Works well in real-world

# Coding Task

```python
 # Dataset

data = {
    "Hours": [1,2,3,4,5,6,7,8],
    "Practice": [5,10,15,20,25,30,35,40],
    "Result": ['Fail','Fail','Fail','Fail','Pass','Pass','Pass','Pass']
}

import pandas as pd
df = pd.DataFrame(data)

X = df[['Hours', 'Practice']]
y = df['Result']
```

```python
# ✅ Step 1: Import

from sklearn.ensemble import RandomForestClassifier
```

```python
# ✅ Step 2: Train Model

model = RandomForestClassifier(n_estimators = 5)  #👉 n_estimators = number of trees
model.fit(X,y)
```

```python
model.predict([[6,30]])
```

## Key Difference
* Decision Tree → 1 tree ❌  
* Random Forest → many trees ✔️

## Final Questions
1. Why is Random Forest better than Decision Tree?

2. What does n_estimators mean?

```python
# Ans 1 :
# Because Random Forest uses multiple trees instead of one,
# which reduces overfitting and gives more accurate and stable predictions.
# Decision Tree → one opinion ❌  
# Random Forest → many opinions ✔️ (better decision)

# Ans 2 : 
# n_estimators = number of trees in the Random Forest
# n_estimators = 5 → 5 trees  
# n_estimators = 100 → 100 trees
```
