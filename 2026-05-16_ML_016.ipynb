# Day 16 — Choosing Best K (Hyperparameter Tuning)

###  Problem
#### In KNN:

* K = 1 → Overfitting ❌  
* K = large → Underfitting ❌


### 👉 So question is:

#### Which K is BEST? 🤔
### Solution: Try Multiple K Values
* We don’t guess ❌
* We test different K values ✔️

### Idea (Very Important)
* Train model with different K  
* Check accuracy  
* Choose best one

#  Coding Task

```python
# Dataset

data = {
    "Hours": [1,2,3,4,5,6,7,8],
    "Result": [0,0,0,1,1,1,1,1]
}
import pandas as pd
df = pd.DataFrame(data)
df
```

```python
# ✅ Step 1: Import

from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
# ✅ Step 2: Split Data

from sklearn.model_selection import train_test_split

X = df[['Hours']]
y = df['Result']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
# ✅ Step 3: Try Different K

for k in range(1,6):
    model = KNeighborsClassifier(n_neighbors = k)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    print(f"K = {k} , Accuracy = {acc}")
```

```python
# 🧠 Best K
# 👉 Multiple K values gave 1.0 accuracy:

# K = 1, 2, 3, 4 ✔️
# 👉 But we should NOT blindly pick K = 1 ❌

# 🎯 Final Answer

# Best K: 3 (or 4 is also fine)

# 🧠 Why NOT K = 1?
# ❓ Your Question:
# Why don’t we always choose K = 1?

# ✅ Correct Answer:
# Because K = 1 can overfit the data.
# It depends too much on a single point and may not generalize well to new data.
```
