## YOUR TASK (STEP BY STEP)

```python
# ✅ Step 1: Create DataFrame

data = {
    "Hours": [1,2,3,4,5,6,7,8],
    "Practice": [10,20,30,40,50,60,70,80],
    "Gender": ["Male","Female","Female","Male","Male","Female","Male","Female"],
    "Marks": [12,22,32,42,52,62,72,82]
}
import pandas as pd

df = pd.DataFrame(data)
df
```

```python
# ✅ Step 2: Handle Encoding
# 👉 Convert Gender → numeric (One-Hot)

encoded_df = pd.get_dummies(df, columns = ['Gender'], dtype = 'int')
encoded_df
```

```python
# ✅ Step 3: Define X and y
# X → all input columns  
# y → Marks

X = encoded_df.drop(columns = 'Marks')
y = encoded_df['Marks']
```

```python
# ✅ Step 4: Train Model
# Use:
# LinearRegression

from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X,y)
```

```python
# ✅ Step 5: Prediction
# Predict for:

# Hours = 9  
# Practice = 90  
# Gender = Male


new_data = pd.DataFrame({'Hours':[9], 'Practice':[90], 'Gender_Female':[0], 'Gender_Male':[1]})
model.predict(new_data)
```
