## 🎯 What is DBSCAN?
**DBSCAN = Density-Based Spatial Clustering of Applications with Noise**

**It groups points that are:**
* Close together (high density) and marks isolated points as:
* Noise / Outliers

```python
# Self Practice

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN

df = pd.DataFrame({"Hours":[1,2,2,3,3,4,8,9,9,10,25]})
print(df)

dbscan = DBSCAN(eps = 2, min_samples = 2)
cluster = dbscan.fit_predict(df)
df["Cluster"] = cluster
print("\n",df)

plt.scatter(df['Hours'], [0]*len(df), df['Cluster'])
```

```python
 # Step 1: Import Libraries

import pandas as pd
import matplotlib.pyplot as plt

from sklearn.cluster import DBSCAN
```

```python
# Step 2: Create Dataset

df = pd.DataFrame({
    "Hours":[1,2,2,3,3,4,8,9,9,10,25]
})

print(df)

# Notice:
# 25
# is far away from the other points.

```

```python
# Step 3: Apply DBSCAN

dbscan = DBSCAN(
    eps=2,
    min_samples=2
)

clusters = dbscan.fit_predict(df)

df["Cluster"] = clusters

print(df)

# Notice:
# -1
# means :  Noise / Outlier
```

## Important Parameters

**eps = 2**

Means:

Maximum distance between points to be considered neighbors.

**min_samples = 2**

Means:

Minimum points needed to form a cluster.

```python
# Step 4: Visualization

plt.scatter(
    df["Hours"],
    [0]*len(df),
    c=df["Cluster"]
)

plt.title("DBSCAN Clustering")
plt.xlabel("Hours")

plt.show()
```

## Why DBSCAN?
**K-Means**
* Must choose K

**DBSCAN**
* No need to choose K Can detect outliers automatically

## In DBSCAN:

* Cluster 0 → Normal Cluster
* Cluster 1 → Another Cluster
* Cluster -1 → Noise / Outlier

Example from your output:

* Hours = 25
* Cluster = -1
* This means the point 25 was too far from other points and was identified as an outlier.

**DBSCAN Quick Recap**
* eps         → Maximum neighbor distance
* min_samples → Minimum points to form a cluster
* -1          → Noise / Outlier
