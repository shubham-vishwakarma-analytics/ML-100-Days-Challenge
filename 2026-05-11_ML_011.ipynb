## Logistic Regression (Let’s Go)

```python
data = {
    "Hours": [1,2,3,4,5,6,7,8],
    "Marks": [10,20,30,40,50,60,70,80]
}
```

```python
# ✅ Step 1: Create DataFrame

import pandas as pd
df = pd.DataFrame(data)
# df
```

```python
# ✅ Step 2: Create Target (IMPORTANT)
# 👉 Convert Marks → Pass/Fail
# Rule:
# Marks < 40 → 0 (Fail)  
# Marks ≥ 40 → 1 (Pass)

df['Result'] = df['Marks'].apply(lambda x : 0 if x < 40 else 1)
```

```python
# ✅ Step 3: Define X and y
# X → Hours  
# y → Result

X = df[['Hours']]
y = df['Result']
```

```python
# ✅ Step 4: Train Model

from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X, y)
```

```python
# ✅ Step 5: Prediction
# Predict for:

# Hours = 3
# Hours = 7

new_data1 = pd.DataFrame({'Hours' : [3]})
print(model.predict(new_data1))
new_data2 = pd.DataFrame({'Hours' : [7]})
print(model.predict(new_data2))
```

## ❗ Final Question (You missed this)
## ❓ Why can’t we use Linear Regression for Pass/Fail?

✅ Correct Answer:

* Because Linear Regression gives continuous output (like 0.3, 0.7),
* but Pass/Fail needs discrete output (0 or 1).
  
👉 Also:

Linear regression can predict values like 1.5 or -0.2 ❌

Not valid for classification
