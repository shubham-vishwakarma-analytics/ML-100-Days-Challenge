#### 🧠 Important Parameters in Random Forest

### 1️⃣ n_estimators
* Number of trees
* More trees → better (but slower)

### 2️⃣ max_depth

Controls depth of each tree

* Small → underfitting ❌  
* Large → overfitting ❌

### 3️⃣ min_samples_split
* Minimum samples required to split a node

### 4️⃣ min_samples_leaf
* Minimum samples required at leaf node

# Coding Task


```python
 # Dataset

data = {
    "Hours": [1,2,3,4,5,6,7,8],
    "Practice": [5,10,15,20,25,30,35,40],
    "Result": ['Fail','Fail','Fail','Fail','Pass','Pass','Pass','Pass']
}

import pandas as pd
df = pd.DataFrame(data)

X = df[['Hours', 'Practice']]
y = df['Result']
```

```python
# ✅ Step 1: Try Different Values

from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

for k in [5,10,15]:
    model = RandomForestClassifier(n_estimators = k)
    model.fit(X, y)
    y_pred = model.predict(X)
    acc = accuracy_score(y, y_pred)
    print(f"Trees = {k} Accuracy = {acc}")

   # 10 is better choice (same accuracy, less computation) ✔️
```

##  Final Questions
1. Why can’t we always use very large n_estimators?

2. What happens if max_depth is too large?

```python
# Ans 1 : 
# We can’t always use very large n_estimators
# because it increases computation time and makes the model slower,
# without significant improvement after a point.


# Ans 2 : 
# Large max_depth → overfitting
```
