## Feature Scaling
###  Concept 
#### Sometimes features have different scales:

**Example:**

* Hours → 1 to 10
* Practice → 10 to 100
  
👉 Model may give more importance to bigger values

🎯 Solution

👉 Feature Scaling

Most common:

StandardScaler

## Task : 1

1. If one feature has very large values and another has small values, what problem can happen?
2. Do you think scaling changes the relationship between data or just the range?

```python
# 1. Model gives more importance to large value features
# → Can lead to biased / inaccurate predictions

# 2. Scaling does NOT change relationship
# It only changes the range (scale) of values
```

## Simple Example

### Before scaling:

* Hours = [1, 2, 3]
* Practice = [10, 20, 30]

### After scaling:

* Hours = [-1, 0, 1]
* Practice = [-1, 0, 1]

## 👉 Relationship is still same:

* Higher Hours → Higher Marks ✔️
* Higher Practice → Higher Marks ✔️
### Only numbers changed, pattern didn’t

🔥 Final Correct Answer

## Scaling does not change relationship, it only changes the range so all features are treated equally.

```python
# Example Dataset
data = {
    "Hours": [1, 2, None, 4, 5],
    "Practice": [10, None, 30, 40, 50],
    "Marks": [15, 30, 45, None, 75]
}
```

```python
# DataFrame
import pandas as pd

df = pd.DataFrame(data)
df = df.fillna(df.mean())
df
```

```python
X = df[['Hours','Practice']]
y = df['Marks']
```

```python
# Scaling
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)

print(X_scaled)
```

## Mini Task
### 👉 Compare:

* X (original)
* X_scaled



```python
print('Mean of X : ', X.mean())
print('Mean of X_scaled : ', X_scaled.mean())
```

```python
print('Standard Deviation of X : ', X.std())
print('Standard Deviation of X_scaled : ', X_scaled.std())
```

# Property (Attribute) of StandarScaler

## 👉 After scaling:

# Mean = 0
# Standard Deviation = 1

# Train Model on Scaled Data
##  Your Task

```python
# train the data on Scaled_X 

from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_scaled, y)
```
