# Day 19 — Distance Metrics in KNN

### KNN works on : Distance between points
* 👉 But question is:
* HOW do we calculate distance? 🤔
# 🎯 Two Main Types

## 1️⃣ Euclidean Distance (Most Common)

#### 👉 Formula:

### √((x1 - x2)² + (y1 - y2)²)

👉 Think:

Straight line distance

#### 🔵 Example

**Points:**

A(2, 3), B(5, 7)

**Distance:**

√((5-2)² + (7-3)²)

= √(9 + 16)

= √25 = 5

## 2️⃣ Manhattan Distance

#### 👉 Formula:

### |x1 - x2| + |y1 - y2|

👉 Think:

City blocks (no diagonal, only straight moves)

#### 🔴 Example

**Same points:**

A(2, 3), B(5, 7)

**Distance:**

|5-2| + |7-3|

= 3 + 4 = 7

## 🧠 Key Difference

**Euclidean → shortest path (diagonal)**  
**Manhattan → step-by-step path (grid)**

### 🔥 When to Use What?

#### Euclidean → default (most cases) ✔️  
**Manhattan → when features are independent / grid-like data**

# Coding Task

```python
 # Dataset

data = {
    "Hours": [1,2,3,4,5,6,7,8],
    "Practice": [10,20,30,40,50,60,70,80],
    "Result": [0,0,0,1,1,1,1,1]
}

import pandas as pd
df = pd.DataFrame(data)
df

X = df[['Hours', 'Practice']]
y = df['Result']

from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
X_scaled = scale.fit_transform(X)

```

```python
 # Use Manhattan Distance in KNN

from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors = 3, metric = 'manhattan')
model.fit(X_scaled, y)

# Notice : manhattan and euclidean Both are in lower case
```

```python
# Prediction

new_data = [[5, 50]]
new_data_scaled = scale.transform(new_data)

print(model.predict(new_data_scaled))
```
