## 🧠 What is Sigmoid Function?
### 👉 It’s the heart of Logistic Regression

## 📈 Formula (don’t fear it):
### 1 / (1 + e^-x)
## 👉 Output will ALWAYS be:

### Between 0 and 1

## 🎯 Why Sigmoid?
#### Because we need:

### Probability (0 to 1)
## Example:

* 0.2 → 20% chance (Fail)  
* 0.8 → 80% chance (Pass)

## 🔵 Shape of Sigmoid Curve


         1 |         ______
           |       /
           |     /
           |   /
           | /
         0 |______________
              -∞    +∞

* 👉 Small input → output near 0
* 👉 Large input → output near 1

## 🧠 How Logistic Regression Works
#### Step 1: Linear equation → (like y = mx + b)

#### Step 2: Apply Sigmoid → convert to probability

#### Step 3: Decision:
    > 0.5 → Class 1 (Pass)
    < 0.5 → Class 0 (Fail)

## 🔥 Example

* If model gives:
* z = 2
* Sigmoid:
* ≈ 0.88 → Pass ✔️

* If:
* z = -2
* Sigmoid:
* ≈ 0.12 → Fail ✔️

## 🛠️ MINI CODING TASK

```python
# 👉 Step 1: Write Sigmoid Function

import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
# 👉 Step 2: Test It

print(sigmoid(-2))
print(sigmoid(0))
print(sigmoid(2))
```

## 🧠 Final Questions
1. Why do we use sigmoid instead of linear output?

2. If sigmoid output = 0.7 → class?

```python
# # Ans 1 :
# Linear regression gives values from -∞ to +∞ (like -10, 50, 100)
# → Not suitable for probability

# Sigmoid converts it into 0 to 1
# → Perfect for probability ✔️

# Ans 2 :
# 0.7 → Pass
```
