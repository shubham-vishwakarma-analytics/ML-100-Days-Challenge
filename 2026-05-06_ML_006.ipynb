## FINAL Mini Project
#### Now we combine everything 👇
### Your Task (Real ML Flow)

```python
# Use the same dirty dataset:

data = {
    "Hours": [1, 2, None, 4, 5],
    "Practice": [10, None, 30, 40, 50],
    "Marks": [15, 30, 45, None, 75]
}
```

```python
# 1. Create DataFrame
# 2. Handle missing values
# 3. Define features
# 4. Train model
# 5. Predict
```

```python
# 1. Create DataFrame

df = pd.DataFrame(data)
df
```

```python
# 2. Handle missing values

df = df.fillna(df.mean())
```

```python
# 3. Define features

X = df[['Hours', 'Practice']]
y = df['Marks']
```

```python
# 4. Train model

from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X, y)
```

```python
# 5. Predict

new_data = pd.DataFrame({
    "Hours": [6],
    "Practice": [60]
})

print(model.predict(new_data))
```

## 👉 Why did we clean data before training, not after?

### We clean data before training because missing values can cause errors and reduce model accuracy. 
### Clean data helps the model learn correct patterns.

##  Important Insight (Don’t Miss This)
