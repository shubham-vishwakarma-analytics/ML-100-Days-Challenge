# Coding Task

```python
 # Dataset

data = {
    "Hours": [1,2,3,4,5,6,7,8],
    "Practice": [5,10,15,20,25,30,35,40],
    "Result": ['Fail','Fail','Fail','Fail','Pass','Pass','Pass','Pass']
}

import pandas as pd
df = pd.DataFrame(data)

X = df[['Hours', 'Practice']]
y = df['Result']

from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(X,y)
```

```python
# ✅ Step 1: Import

from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
```

```python
# ✅ Step 2: Plot Tree

plt.figure(figsize=(15,6))
plot_tree(model, feature_names=['Hours', 'Practice'], class_names=['Fail', 'Pass'], filled=True)
plt.show()
```

```python
# ✅ Step 2: Plot Tree

plt.figure(figsize=(15,6))
plot_tree(model, feature_names=['Hours', 'Practice'], class_names=['Fail', 'Pass'])
plt.show()
```
