## What is Apriori?

Apriori is an algorithm used to find:

Association Rules between items in a dataset.

**Example:**

* Bread → Butter
* Milk → Bread
* Tea → Biscuits

```python
# Step 1: Install Library

!pip install mlxtend
```

```python
# Step 2: Import Libraries

import pandas as pd

from mlxtend.frequent_patterns import (
    apriori,
    association_rules
)
```

```python
 # Step 3: Create Dataset

data = {
    'Bread':[1,1,0,1,1],
    'Butter':[1,1,1,0,1],
    'Milk':[1,0,1,1,1]
}

df = pd.DataFrame(data)

print(df)
```

```python
# Step 4: Find Frequent Itemsets

frequent_items = apriori(
    df,
    min_support=0.4,
    use_colnames=True
)

print(frequent_items)
```

## What is Support?

**Support means:**

How frequently an item appears in the dataset.

**Example:**

Bread appears in 4 out of 5 rows

Support = 4/5 = 0.8

```python
# Step 5: Generate Rules

rules = association_rules(
    frequent_items,
    metric="confidence",
    min_threshold=0.5
)

print(rules)
```

 ## What is Confidence?

Confidence means:

How often Y occurs when X occurs.

**Example:**

Bread → Butter

If:

80% of Bread buyers also buy Butter then:

Confidence = 0.8


## Important Metrics

### Support : Frequency of item
### Confidence : Reliability of rule
### Lift : Strength of relationship
