## 🧠 First Answer This
### ❓ Situation:
* Model says: “Person is healthy”
* But actually: Person is sick

### 👉 This is:

* ❌ VERY DANGEROUS mistake
### Because:

* Patient doesn’t get treatment
* Can become serious

## In classification, we have 2 types of mistakes:

### 🔴 Type 1: False Positive (FP)
* Model says: Sick  
* Actually: Healthy
* 👉 Not too dangerous (extra check can fix)

### 🔵 Type 2: False Negative (FN)
* Model says: Healthy  
* Actually: Sick
* 👉 🚨 VERY dangerous

## 🧠 Precision vs Recall

### ✅ Precision
* 👉 “How many predicted positives are actually correct?”
* Precision = Correct Positive / Total Predicted Positive
* 💡 Focus : Avoid False Positives
### ✅ Recall
* 👉 “How many actual positives did we catch?”
* Recall = Correct Positive / Total Actual Positive
* 💡 Focus : Avoid False Negatives (VERY IMPORTANT in medical)

### Easy Memory Trick
* ##  Precision → Be Accurate  
* ##  Recall → Don’t Miss Important Cases

## 🧠 Real-Life Examples
#### 📧 Spam Detection
* Precision important (don’t mark normal email as spam)

#### 🏥 Disease Detection
* Recall important (don’t miss sick people)
