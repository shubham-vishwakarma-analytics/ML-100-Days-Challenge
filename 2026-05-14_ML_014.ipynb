## 4 Important Terms

### ✅ 1. True Positive (TP)
* Model: Sick  
* Actual: Sick ✔️
### ❌ 2. False Positive (FP)
* Model: Sick  
* Actual: Healthy ❌
### ❌ 3. False Negative (FN)
* Model: Healthy  
* Actual: Sick ❌ (Dangerous)
### ✅ 4. True Negative (TN)
* Model: Healthy  
* Actual: Healthy ✔️
