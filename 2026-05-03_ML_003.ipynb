## Concept

Till now:

👉 You trained and tested on same data

❌ Problem:

##### Model may memorize, not learn

🔥 Solution

👉 Split data into:

#### Training data → model learns

#### Testing data → model is evaluated

* 📦 Example
* Total Data → 100%

* Train → 80%
* Test → 20%

#### : 1. Why can’t we test on same data we train on?

**👉 If we use same data:**

* Model may memorize data (not actually learn)

* It will give perfect accuracy (fake confidence)

* ❌ But in real-world → it may fail on new data

* ✔️ So we separate data to check real performance

#### 2. What is the purpose of test data?
**👉 Test data is used to:**

* Check how well model performs on unseen data

* Measure real accuracy



## Tasks

```python
# Dataset
import pandas as pd
data = pd.DataFrame({'Hours' : [1,2,3,4,5,6,7] , 'Marks' : [12,24,36,48,60,72,84]})
# data
```

```python
# Split the data
from sklearn.model_selection import train_test_split

x = data[["Hours"]]
y = data["Marks"]

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred  = model.predict(X_test)
y_pred
```

## Mini Task
##### Actual vs Predicted

```python
print(f"Actual : {y_test}")
print(f"Predicted : {y_pred}")
```
